# Jump Sequence Indexing

Implement `solve(jump_sequence: tuple[int, ...], start_index: int) -> bool`.

Each tuple element is the exact distance you can move left or right from that index. Return `True` if any reachable index contains `0`; otherwise return `False`.

```python
from collections import deque


def solve(jump_sequence: tuple[int, ...], start_index: int) -> bool:
    """Return True when a zero-valued tuple position is reachable."""
    if not 0 <= start_index < len(jump_sequence):
        return False

    visited: set[int] = set()
    queue: deque[int] = deque([start_index])

    while queue:
        index = queue.popleft()
        if index in visited:
            continue

        visited.add(index)
        jump = jump_sequence[index]
        if jump == 0:
            return True

        for next_index in (index - jump, index + jump):
            if 0 <= next_index < len(jump_sequence) and next_index not in visited:
                queue.append(next_index)

    return False
```

```python
test_cases = [
    ((4, 9, 2, 100, 0), 2, True),
    ((1, 0, 9, 2, 100, 1), 3, True),
    ((0, 9), 1, False),
    ((1, 1, 0), 0, True),
    ((), 0, False),
    ((4, 9, 2, 100, 2), 2, False),
]

for jump_sequence, start_index, expected in test_cases:
    actual = solve(jump_sequence, start_index)
    assert actual is expected, (jump_sequence, start_index, expected, actual)

"All test cases passed."
```
